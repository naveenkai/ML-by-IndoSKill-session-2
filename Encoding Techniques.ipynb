```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

## fit -- for fitting
## transform -- for transforming
## fit_transform -- for both

data['sex']=le.fit_transform(data['sex'])
```

```python
data['smoker']=le.fit_transform(data['smoker'])
```

## Feature Mapping

```python
## LabelEncoding will not rank

# for ordinal data -- for ranked column

data1 = pd.read_csv('bigmart.csv')
print(data1['Outlet_Size'].value_counts())
data1['Outlet_Size'].fillna(data1['Outlet_Size'].mode()[0], inplace = True)
data1['Outlet_Size'] = data1['Outlet_Size'].replace(('High','Medium','Small'),(3,2,1))
data1.head()
```
